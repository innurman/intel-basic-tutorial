## Savefig 명령어를 사용하여 차트 저장하기

`savefig` 명령어를 사용하여 원하는 포맷(png, jpg, pdf 등..)으로 저장 할 수 있습니다. 

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
```

## png파일로 저장하기 

savefig 명령어의 첫번째 입력 값으로 저장 하는 파일 포맷을 png로 설정해 줍니다. 
- `pad_inches = 1 `로 여백사이즈를 조절할 수 있습니다. 
-  `transparent`로 배경을 없애 저장 할 수 있습니다.

```python
plt.axis("equal")
plt.pie(exp_vals,
        labels=exp_labels,
        shadow=True,
        autopct='%1.1f%%',
        radius=1.5,
        explode=[0,0,0,0.1,0.2],
        startangle=45)

plt.savefig("img/piechart.png",
            bbox_inches="tight",
            pad_inches=1,
            transparent=True)
plt.show()
```

##  pdf 로 저장하기 

savefig 명령어의 첫번째 입력 값으로 저장 하는 파일 포맷을 pdf로 설정해 줍니다. 

```python
plt.axis("equal")
plt.pie(exp_vals,
        labels=exp_labels,
        shadow=True,
        autopct='%1.1f%%',
        radius=1.5,
        explode=[0,0,0,0.1,0.2],
        startangle=45)

plt.savefig("img/piechart.pdf",
            bbox_inches="tight",
            pad_inches=10,
            transparent=True)

plt.show()
```
